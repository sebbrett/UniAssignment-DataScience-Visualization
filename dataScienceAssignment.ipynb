```python
# prompt: visualize skew in multiple plots

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the dataframe 'df' is already loaded as in your provided code.

# Select numerical columns for skew analysis (exclude binary columns)
numerical_cols = ['age', 'education', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']

# Create subplots for each numerical feature
num_plots = len(numerical_cols)
n_rows = (num_plots + 2) // 3  # Calculate rows for 3 plots per row
fig, axes = plt.subplots(nrows=n_rows, ncols=min(num_plots, 3), figsize=(15, 5 * n_rows))

# Flatten axes array if necessary
if num_plots == 1:
  axes = [axes]
elif num_plots > 1:
  axes = axes.flatten()

for i, col in enumerate(numerical_cols):
    # Distribution plot
    sns.histplot(df[col], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

    # Calculate and display skewness"
    skew_val = df[col].skew()
    axes[i].text(0.05, 0.95, f'Skewness: {skew_val:.2f}', transform=axes[i].transAxes, fontsize=10, verticalalignment='top')


# Remove any empty subplots if there are less than 3*n_rows plots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#### Question 2.2: What are the mean systolic BP among subjects at risk according to age group

```python
# prompt: mean sysBP of the tenyearchd =1 according to age group, then compare with mean of all sysBP, use sns

import matplotlib.pyplot as plt
# Calculate the mean sysBP for each age group where TenYearCHD == 1
mean_sysBP_by_age_group = df[df['TenYearCHD'] == 1].groupby('age_group')['sysBP'].mean()

# Calculate the overall mean sysBP
overall_mean_sysBP = df['sysBP'].mean()

# Create a bar plot to compare mean sysBP by age group with the overall mean
plt.figure(figsize=(10, 8))
sns.barplot(x=mean_sysBP_by_age_group.index, y=mean_sysBP_by_age_group.values, color='#a3c4f3', label='Mean sysBP (TenYearCHD=1)')
plt.axhline(y=overall_mean_sysBP, color='red', linestyle='--', label=f'Overall Mean sysBP: {overall_mean_sysBP:.2f}')
plt.xlabel('Age Group')
plt.ylabel('Mean sysBP')
plt.title('Mean sysBP of Subjects at Risk (TenYearCHD=1) by Age Group vs. Overall Mean')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

Inference 2.1: The sysBP increase proportionally to age, and since the correlation of systolic blood pressure (sysBP) and diastolic blood pressure (diaBP) is positively correlated, it can be deducted that the attribute diaBP has similar trend with sysBP.
